#Data preparation  
1.First, we convert the DICOM file to jpg file for data input.

```python
import os
import matplotlib.pyplot as plt
import pydicom
import png
from pydicom.data import get_testdata_files
import scipy.misc

folder = '../all/stage_2_train_images'
filenames = os.listdir(folder)

# in_path = '0a0f91dc-6015-4342-b809-d19610854a21.dcm'
# out_path = '0a0f91dc-6015-4342-b809-d19610854a21.jpg'

for name in filenames:
    ds = pydicom.read_file(folder+"/"+name)  #read dcm file
    img = ds.pixel_array  # read image array information
    scipy.misc.imsave("input_jpg/"+name+".jpg",img) #output jpg file
```
